1. filter out the missing proteins with missing in more than 30% samples in either <strong> control</strong>
 or <strong>asthma</strong> samples

- based on the excel sheets we have 2, 'before' and 'after'
- and then we have the the 'asthma' and 'control' columns in each of these sheets, representing the samples 

```python
def catagorizer(df):
    asthma_columns = []
    control_columns = []
    for col in df.columns:
        if 'control' in col:
            control_columns.append(col)
        elif 'asthma' in col:
            asthma_columns.append(col)
        elif col == 'Protiens':
            proteins_column = df[col]
            
    df_asthma = df[asthma_columns]
    df_control = df[control_columns]
    
    if 'Protiens' in df.columns:
        df_asthma = df_asthma.assign(Protiens=df['Protiens'])
        df_control = df_control.assign(Protiens=df['Protiens'])
    else:
        df_asthma = df_asthma.assign(Protiens=None)
        df_control = df_control.assign(Protiens=None)
    
    return df_asthma, df_control
```

```python
# run the function catagorizer here on both before and after datasets
df_before_asthma, df_before_control = catagorizer(df_before)
#after
df_after_asthma, df_after_control = catagorizer(df_after)
```

```python
# we have specified the path: categorized_dir = config['datasets']['categorized_dir']

def filter_proteins(df, presence_threshold, output_dir, filename):
    # Calculate the minimum number of non-NA values required
    min_presence_count = presence_threshold * df.shape[1]
    
    # Drop rows with less than the calculated threshold of non-NA values
    filtered_df = df.dropna(thresh=min_presence_count)
    
    # Save the filtered dataframe to the specified directory
    output_path = os.path.join(output_dir, filename)
    filtered_df.to_csv(output_path, index=False)
    
    return filtered_df

# Adjust the threshold as needed - here it is 0.7 for 70%
df_before_control_filtered = filter_proteins(df_before_control, 0.7, categorized_dir, 'df_before_control_filtered.csv')
df_before_asthma_filtered = filter_proteins(df_before_asthma, 0.7, categorized_dir, 'df_before_asthma_filtered.csv')
df_after_control_filtered = filter_proteins(df_after_control, 0.7, categorized_dir, 'df_after_control_filtered.csv')
df_after_asthma_filtered = filter_proteins(df_after_asthma, 0.7, categorized_dir, 'df_after_asthma_filtered.csv')
```
